```rust
fn main() {
    let condition = true;
    let number = if condition { 5 } else { 6 };

    println!("The value of number is: {number}");
}
main();
```

Смысл в том, что if тут такой же, как обычно, просто в 1 строку. У нас выражение `if true` -> 5, `else` -> 6; 
condiiton по определению true. If здесь показан как выражение.
