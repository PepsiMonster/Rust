### Кортежи

В кортеже могут быть любые данные, у них есть номера
Прописывать типа данных, как в примере ниже не обязательно

Tup от слова tupple - кортеж 

```rust
fn main() {
    let tup: (i32, f64, u8) = (500, 6.4, 1);
}
main();
```

Получение отдельного элемента кортежа

```rust
fn main() {
    let tup = (500, 6.4, 1); // создали кортеж связанный с переменной tup

    let (x, y, z) = tup; // tup превращается в 3 отдельные переменные (деструктуризация)

    println!("The value of y is: {y}");
}
main();
```

```rust
fn main() {
    let x: (i32, f64, u8) = (500, 6.4, 1);

    let five_hundred = x.0;

    let six_point_four = x.1;

    let one = x.2;

    println!("{one}");
    // return (five_hundred, six_point_four, one);
}
main();
```

### Массивы (array)

Каждый элемент должен иметь один и тот же тип

```rust
fn main() {
    let a = [1, 2, 3, 4, 5];
    let num = a.0;
}
main();
```

Массивы имеют фиксированное кол-во элементов после создания. Удобно, если мы точно знаем, что число элементов не будет меняться

```rust
let months = ["January", "February", "March", "April", "May", "June", "July",
              "August", "September", "October", "November", "December"];
```

Тип массива записывается [`тип данных`:`число элементов`]

```rust
let a: [i32; 5] = [1, 2, 3, 4, 5];
```

Массив можно заполнить повторяющимся элементом

```rust
let a = [3; 5];
println!("{:?}", a);  // Необходимо :? чтобы напечатать массив
```

```rust
let a = [3; 5];
a // a)
```

Доступ к элементам массива получается через индексацию

```rust
fn main() {
    let a = [1, 2, 3, 4, 5];

    let first = a[0];
    let second = a[1];
    println!("{}, {}", second, first);

}
main();
```
