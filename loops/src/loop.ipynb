```rust
fn main(){
    let mut count = 3;
    loop {
        println!("remaining count: {}", count);
        if count == 0 {
            break println!("Liftoff");
        }
        else {
            count -= 1;
        }
    }
}
main();
```

Чтобы не использовать сложные if и else в таких попсовых случаях, сделали частный случай while. Удобно. 

```rust
fn main() {
    let mut number = 3;

    while number != 0 {
        println!("{number}!");

        number -= 1;
    }

    println!("LIFTOFF!!!");
}
main();
```

```rust
fn main() {
    let a = [10, 20, 30, 40, 50];   // Если убрать 1 переменную
    let mut index = 0; 

    while index < 5 {               // Придется ее и тут поменять
        println!("the value is: {}", a[index]);

        index += 1;
    }
}
main();
```

Чтобы не менять каждый раз индекс массива в 2х местах, есть цикл for. Частный случай while.

```rust
let a = [10, 20, 30, 40, 50];  // Это коробка с 5 элементами

for element in a {              // "Для каждого элемента внутри a"
    println!("{element}");      // "Напечатай этот элемент"
}
```

```rust
// Rust: тоже самое!
for i in 0..5 {
    println!("{i}");
    let i = 100; // Создаем новую переменную с тем же именем
    println!("Новый i: {i}");
    // На следующей итерации i снова будет из диапазона
}
```

```rust
fn main() {
    for number in (1..4).rev() {    // (1..4) значит от 1 (включительно) до 4 (не включительно)
        println!("{number}!");      // круглые скобки тут нужны только для .rev(), можно просто 
                                    // for number in 1..4 {}
    }
    println!("LIFTOFF!!!");
}
main();
```
